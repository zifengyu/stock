```python
import pandas as pd

pd.options.display.float_format = '{:,.2f}'.format
```

```python
profit = 32
rate5 = 0.1
rate10 = 0.05
pe10 = 5
dividend = 0.3
discount = 0.1

rate = [(1 + rate5) ** i for i in range(1, 6)] + [((1 + rate5) ** 5) * ((1 + rate10) ** i) for i in range(1, 6)]
df = pd.DataFrame([profit * r for r in rate], columns =['Profit'])
df['Dividened'] = df['Profit'] * dividend
df['Present'] = [(1 + discount) ** i for i in range(1, 11)]
fdc = df['Present'].iloc[-1]
fprofit = df['Profit'].iloc[-1]
df['Present'] = df['Dividened'] / df['Present']
sum = df['Present'].sum() + pe10 * fprofit / fdc

display(df)
print("final: %.2f %.2f" % (fprofit, fdc))
print("total: %.2f" % sum)
```
