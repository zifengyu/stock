```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import script.data as data
```

```python
df = data.get_IS()
```

```python
year = 3
delta = 0
grouped = df.groupby('股票代码')
tickers = []
for t, group in grouped:
    if len(group) > year+delta and group['归属于母公司所有者的净利润'].pct_change(periods=-1)[delta:delta+year].min() > 0.25:
        last = group
        if not t.startswith('A') and not t.startswith('300'):
            tickers.append(t)
print(len(tickers))
```
